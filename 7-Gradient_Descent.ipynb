```python
import numpy as np
import matplotlib.pyplot as plt

def f(x):
    return x**2

def df(x):
    return 2 * x

learning_rate = 0.1
n_iters = 20

x = 3.0

xs = []
fs = []

for i in range(n_iters + 1):
    xs.append(x)
    fs.append(f(x))
    
    grad = df(x)
    
    x = x - learning_rate * grad

x_plot = np.linspace(-4, 4, 400)
y_plot = f(x_plot)

plt.figure(figsize=(6, 3))
plt.plot(x_plot, y_plot, label="f(x) = x^2")
plt.scatter(xs, fs, marker='o', label="Gradient Descent steps")
plt.plot(xs, fs, linestyle='--')
plt.scatter(xs[-1], fs[-1], s=80, zorder=3)
plt.title("Gradient Descent on f(x) = x^2")
plt.xlabel("x")
plt.ylabel("f(x)")
plt.legend()
plt.grid(True)
plt.show()
```
